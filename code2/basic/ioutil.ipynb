```python
import numpy as np
import cv2
import pickle
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../opencv/a.JPG')
print(type(img))
```

```python
bt = pickle.dumps(img)
result = pickle.loads(bt)
plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import base64
bs64 = base64.b64encode(bt)
tt = base64.b64decode(bs64) 
img = pickle.loads(tt)
print(type(img))
```

```python
with open("../opencv/a.JPG","rb") as f:  
    # b64encode是编码，b64decode是解码  
#     base64_data = base64.b64encode(f.read())  
    # base64.b64decode(base64data)  
#     print(base64_data)
#     result2 = base64.b64decode(base64_data)
#     print(type(result2))
    nparr = np.fromstring(f.read(), np.uint8)
    img_np = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    # print(img_np.shape)
    cv2.imshow("test",img_np)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
